```python
# Load the CSV data
df1 = pd.read_csv('Population_different_countries.csv')
population_brazil = df1.loc[df1['Country'] == 'Brazil', 'Population (Millions)'].values[0]
df1 = df1[(df1['Population (Millions)'] <= population_brazil)]

# Create the figure with two polar subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10), subplot_kw={'projection': 'polar'})
countries = df1['Country']
populations = df1['Population (Millions)']

# Normalize population for consistent scaling
max_population = max(populations)
normalized_population = populations / max_population

# Define properties for the radial plot
num_countries = len(countries)
angles = np.linspace(0, 2 * np.pi, num_countries, endpoint=False)
bar_width = 2 * np.pi / num_countries - 0.05  # Create gaps between slices
inner_radius = 0.9  # Start bars from this radius

# Plot each slice as a bar in the radial chart
bars = ax1.bar(
    angles,
    normalized_population,  # Height starts from inner_radius
    width=bar_width,
    bottom=inner_radius,
    color='tab:orange',
    edgecolor='white',  # Create gap effect
    linewidth=2
)

# Add a white circle at the center to make the inner radius appear empty
ax1.fill_between(np.linspace(0, 2 * np.pi, 100), 0, inner_radius, color='white', zorder=10)

# Place country names at the end of each bar in a radial manner
for i, (bar, angle, name, population) in enumerate(zip(bars, angles, countries, populations)):
    height = bar.get_height()
    rotation = np.degrees(angle)

    # Ensure text is upright
    alignment = 'center'
    if 90 < rotation < 270:
        rotation += 180
        alignment = 'center'

    # Place the text at the end of each bar
    ax1.text(
        angle,
        height + 0.3 + inner_radius,  # Slightly outside the bar
        f'{name} ({population})',
        ha=alignment,
        va='center',
        rotation=rotation,
        rotation_mode='anchor',
        fontsize=12
    )

# Set properties for better visualization
ax1.set_yticks([])  # Remove radial grid lines
ax1.set_xticks([])  # Remove ticks on bars

# Hide the outer circle (polar frame)
ax1.spines['polar'].set_visible(False)

# Alternatively, hide all gridlines and the frame
ax1.grid(False)

# Title for each subplot
ax1.set_title(title, fontsize=14)






df2 = pd.read_csv('Brazil_population.csv')
# Create the second radial chart
year = df2['Year']
populations = df2['Population (Millions)']

# Normalize population for consistent scaling
max_population = max(populations)
normalized_population = populations / max_population

# Define properties for the radial plot
num_year = len(year)
angles = np.linspace(0, 2 * np.pi, num_year, endpoint=False)
bar_width = 2 * np.pi / num_year - 0.05  # Create gaps between slices
inner_radius = 0.6  # Start bars from this radius

# Plot each slice as a bar in the radial chart
bars = ax2.bar(
    angles,
    normalized_population-0.5,  # Height starts from inner_radius
    width=bar_width,
    bottom=inner_radius,
    color='tab:orange',
    edgecolor='white',  # Create gap effect
    linewidth=2
)

# Add a white circle at the center to make the inner radius appear empty
ax2.fill_between(np.linspace(0, 2 * np.pi, 100), 0, inner_radius, color='white', zorder=10)

# Place country names at the end of each bar in a radial manner
for i, (bar, angle, name, population) in enumerate(zip(bars, angles, year, populations)):
    height = bar.get_height()
    rotation = np.degrees(angle)

    # Ensure text is upright
    alignment = 'center'
    if 90 < rotation < 270:
        rotation += 180
        alignment = 'center'

    # Place the text at the end of each bar
    ax2.text(
        angle,
        height + 0.2 + inner_radius,  # Slightly outside the bar
        f'{name} ({population})',
        ha=alignment,
        va='center',
        rotation=rotation,
        rotation_mode='anchor',
        fontsize=12
    )

# Set properties for better visualization
ax2.set_yticks([])  # Remove radial grid lines
ax2.set_xticks([])  # Remove ticks on bars

# Hide the outer circle (polar frame)
ax2.spines['polar'].set_visible(False)

# Alternatively, hide all gridlines and the frame
ax2.grid(False)

# Title for each subplot
ax2.set_title(title, fontsize=14)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# Load the CSV data
df = pd.read_csv('Population_different_countries.csv')
population_brazil = df.loc[df['Country'] == 'Brazil', 'Population (Millions)'].values[0]
df = df[(df['Population (Millions)'] <= population_brazil)]
df = df.sort_values(by='Population (Millions)', ascending=True)

# Extract country names and populations
countries = df['Country']
populations = df['Population (Millions)']

# Normalize population based on 600 million corresponding to 3/4 of the circle
max_population = 400
normalized_population = populations / max_population

# Define properties for the radial plot
num_countries = len(countries)
start_angle = -np.pi / 2   # Start from the third quadrant
arc_span = -3 * np.pi / 2  # Clockwise rotation through 3/4 of the circle
bar_width = 15              # Thickness of each bar
initial_radius = 60        # Larger starting radius for better spacing
radius_increment = 2       # Small, consistent increment for each arc
gray_color = '#c0c0c0'    # Color of the base arcs
blue_color = 'tab:blue'    # Color of the population bars
text_color = 'black'        # Color of the country names
before = 6

# Create the figure and polar subplot
fig, ax = plt.subplots(figsize=(9.5, 9.5), subplot_kw={'projection': 'polar'})

for i in range(before):
  theta_start = start_angle
  theta_end = start_angle + arc_span
  theta = np.linspace(theta_start, theta_end, 100)

  # Fixed initial radius + consistent increment
  r = np.full_like(theta, initial_radius + (i * radius_increment))

  # Plot the base arc in gray
  ax.plot(theta, r, color="white", linewidth=bar_width)

# Plot each base arc and population bar
for i, (country, norm_pop, population) in enumerate(zip(countries, normalized_population, populations)):
    # Calculate the starting and ending angles for the base arc
    theta_start = start_angle
    theta_end = start_angle + arc_span
    theta = np.linspace(theta_start, theta_end, 100)

    # Fixed initial radius + consistent increment
    r = np.full_like(theta, initial_radius + ((i+before) * radius_increment))

    # Plot the base arc in gray
    ax.plot(theta, r, color=gray_color, linewidth=bar_width)

    # Plot the population bar in blue on top of the gray arc
    pop_theta_end = theta_start + (arc_span * norm_pop)
    pop_theta = np.linspace(theta_start, pop_theta_end, 100)
    ax.plot(pop_theta, r, color=blue_color, linewidth=bar_width)

    # Plot the white slab at the end of the blue bar
    slab_theta_start = pop_theta_end
    slab_theta_end = slab_theta_start - np.pi/1000  # Small angle for the white slab
    slab_theta = np.linspace(slab_theta_start, slab_theta_end, 100)
    ax.plot(slab_theta, r, color='white', linewidth=bar_width)

    # Place the country name linearly outside the start axis
    ax.text(
        theta_start,              # Keep the same angle
        r[0],      # Constant radial distance for all labels
        f'     {country}',
        ha='left',               # Align text to the right
        va='center',
        fontsize=10,
        color=text_color,
        rotation=0                # Horizontal text
    )

# Set properties for better visualization
ax.set_yticks([])  # Remove radial grid lines
ax.set_xticks([])  # Remove ticks on bars
ax.spines['polar'].set_visible(False)  # Hide the outer circle

# Title and display
plt.title('3/4 Ring Radial Chart of Population by Country', fontsize=16, fontweight='bold', pad=40)
plt.show()
```

```python
# Load the dataset
df = pd.read_csv('/content/poverty_population.csv')

df = df.sort_values(by='Poverty',ascending=True).reset_index(drop=True)

from matplotlib.colors import LinearSegmentedColormap
countries = ['Brazil', 'Indonesia', 'Nigeria', 'China', 'India']
# Calculate the total population and percentage contribution
total_population = df['Population'].sum()
df['Population %'] = (df['Population'] / total_population) * 100

# Calculate cumulative percentage for positioning on the x-axis
df['Cumulative Population %'] = df['Population %'].cumsum().shift(fill_value=0)

# Create figure and axis
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(18, 6))


# plot left
# Define custom color map for poverty
colors = ["#8B4513", "#ff7700", "#FFFFE0", "#FFFFE0", "#FFFFE0", "#edd37b"]  # Brown to Light Yellow
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

# Normalize poverty data for color mapping
norm = plt.Normalize(df['Poverty'].min(), df['Poverty'].max())

# Plot the Marimekko chart
for i, row in df.iterrows():
    # Calculate width and position for each country
    width = row['Population %']
    left = row['Cumulative Population %']
    height = row['Poverty']

    # Get color for current poverty value
    if (i == 0):
      color = cmap(norm(height))
    else:
      color = cmap(norm(height+5))

    # Plot rectangle for each country
    ax1.bar(left, height, width=width, align='edge', color=color, edgecolor=color)

    # Add text label for country name
    if row['Country'] in countries:
      ax1.text(left + width/2, height + 1, row['Country'], ha='right', fontsize=10, rotation=0)

# Customize the plot
ax1.set_xlabel('Population Percentage')
ax1.set_ylabel('Poverty Percentage')
ax1.set_title('Hight Population, High Poverty \nPercent of people with less than $5.20/day', fontweight='bold', pad = 30)

# Set x-axis limits and labels
ax1.set_xlim(0, 100)
ax1.set_xticks(np.arange(0, 101, 10))
ax1.set_yticks(np.arange(0, df['Poverty'].max()+10, 10))

# Grid and style adjustments
ax1.grid(axis='y', linestyle=custom_linestyle, alpha=0.5)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Set x-axis labels as percentages
ax1.set_xticklabels([f'{int(x)}%' for x in ax1.get_xticks()])
# # Set y-axis labels as percentages
ax1.set_yticklabels([f'{int(y)}%' for y in ax1.get_yticks()])



# Right chart
# Plot the Marimekko chart
for i, row in df.iterrows():
    # Calculate width and position for each country
    width = row['Population %']
    left = row['Cumulative Population %']
    height = row['Poverty']

    # Color condition: Blue for Brazil and China, Orange for others
    if row['Country'] in ['Brazil', 'China']:
        color = 'tab:blue'
    else:
        color = 'tab:orange'

    # Plot rectangle for each country
    ax2.bar(left, height, width=width, align='edge', color=color, edgecolor=color)

    # Add text label for country name
    if row['Country'] in countries:
      ax2.text(left + width/2, height + 1, row['Country'], ha='right', fontsize=10, rotation=0)

# Customize the plot
ax2.set_xlabel('Population Percentage')
ax2.set_ylabel('Poverty Percentage')
ax2.set_title('Hight Population, High Poverty \nPercent of people with less than $5.20/day', fontweight='bold', pad = 30)

# Set x-axis limits and labels
ax2.set_xlim(0, 100)
ax2.set_xticks(np.arange(0, 101, 10))
ax2.set_yticks(np.arange(0, df['Poverty'].max()+10, 10))

# Grid and style adjustments
ax2.grid(axis='y', linestyle=custom_linestyle, alpha=0.5)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# Set x-axis labels as percentages
ax2.set_xticklabels([f'{int(x)}%' for x in ax2.get_xticks()])

# # Set y-axis labels as percentages
ax2.set_yticklabels([f'{int(y)}%' for y in ax2.get_yticks()])

plt.show()
```
